# Measure Dynamic Functional Connectivity

## Initialize Environment

```python
try:
    %load_ext autoreload
    %autoreload 2
    
except:
    print 'NOT IPYTHON'

from __future__ import division
from IPython.display import display

import os
import sys
import glob
import json

import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt

import dyne

path_CoreData = '/data/jag/akhambhati/hoth_research/CoreData/IEEG_Neocortical'
path_PeriphData = '/data/jag/akhambhati/hoth_research/PeriphData/ds-VCR_PushPull'
path_ExpData = path_PeriphData + '/e01-Dyne_FuncNetw'

for path in [path_CoreData, path_PeriphData, path_ExpData]:
    if not os.path.exists(path):
        print('Path: {}, does not exist'.format(path))
        os.makedirs(path)
```

## Generate List of Data

```python
subjs = [full_subj_path.split('/')[-1]
         for full_subj_path in glob.iglob('{}/HUP*'.format(path_CoreData))]

subj_ids = {}
proc_path = []
for subj in subjs:
    event_path = glob.glob('{}/{}/*.mat'.format(path_CoreData, subj))
    subj_ids[subj] = {'event_path': [],
                      'event_name': []}
    
    for path in event_path:
        subj_ids[subj]['event_path'].append(path)
        subj_ids[subj]['event_name'].append(path.split('/')[-1].split('.')[0])
```

## Design and Run DyNe Pipeline

### Generate Pipe Definitions (JSON Files)

```python
# Coherence Definitions
coh = {'AlphaTheta': [5., 15.],
       'Beta': [15., 25.],
       'LowGamma': [30., 40.],
       'HighGamma': [95., 105.]}

for subj, events in subj_ids.items():
    for ev_name, ev_path in zip(events['event_name'],
                                events['event_path']):
        for band in coh.iterkeys():

            # Define the pipe definitions
            pipe_defs = {
                "SOURCE": {
                    "PIPE_NAME": "Raw_ECoG",
                    "PIPE_MODULE": "dyne.interface.offline",
                    "PIPE_CLASS": "MATSignal",
                    "PIPE_PARAM": {
                        "signal_path": ev_path,
                        "win_len": 1.0,
                        "win_disp": 1.0
                    }
                },

                "FLOW": [
                    {
                        "PIPE_NAME": "CARFilter",
                        "PIPE_MODULE": "dyne.preproc.filters",
                        "PIPE_CLASS": "CommonAvgRef",
                        "PIPE_PARAM": {}
                    },

                    {
                        "PIPE_NAME": "MTCoh",
                        "PIPE_MODULE": "dyne.adjacency.coherence",
                        "PIPE_CLASS": "MTCoh",
                        "PIPE_PARAM": {
                            "time_band": 5.,
                            "n_taper": 9,
                            "cf": coh[band]
                        }
                    },
                    
                    {
                        "PIPE_NAME": "Synchronizability",
                        "PIPE_MODULE": "dyne.globaltopo.centrality",
                        "PIPE_CLASS": "Synchronizability",
                        "PIPE_PARAM": {}
                    },
                    
                    {
                        "PIPE_NAME": "VirtualResection",
                        "PIPE_MODULE": "dyne.nodetopo.centrality",
                        "PIPE_CLASS": "SyncCentral",
                        "PIPE_PARAM": {}
                    },

                    {
                        "PIPE_NAME": "DegreeCentrality",
                        "PIPE_MODULE": "dyne.nodetopo.centrality",
                        "PIPE_CLASS": "DegrCentral",
                        "PIPE_PARAM": {}
                    },
                    
                    {
                        "PIPE_NAME": "CachePipeline",
                        "PIPE_MODULE": "dyne.logger.cache",
                        "PIPE_CLASS": "SaveHDF",
                        "PIPE_PARAM": {
                            "path": "{}/{}.{}.dyne_output.hdf".format(path_ExpData,
                                                                      ev_name, 
                                                                      band)
                        }
                    }
                ],

                "LOG": {
                    "PATH": "{}/{}.{}.dyne_log.csv".format(path_ExpData,
                                                           ev_name,
                                                           band)
                }
            }
            json.dump(pipe_defs, open("{}/{}.{}.dyne_pdef.json".format(path_ExpData,
                                                                       ev_name,
                                                                       band),
                                      'w'))
```

### Generate Pipeline Definitions (JSON File)

```python
pipeline_def = {
    "Raw_ECoG": [
        "CARFilter"],

    "CARFilter": [
        "MTCoh"],

    "MTCoh": [
        "Synchronizability",
        "VirtualResection",
        "DegreeCentrality",
        "CachePipeline"],

    "Synchronizability": [
        "CachePipeline"],

    "VirtualResection": [
        "CachePipeline"],

    "DegreeCentrality": [
        "CachePipeline"],

    "CachePipeline": [
        "None"]
}
json.dump(pipeline_def, open("{}/ecog_coh_pipeline.json".format(path_ExpData),
                             "w"))
```

### Run the Pipeline (in Parallel) 
*__WARNING: Will Delete Existing Output__*

```python
# Remove all existing output (retains pipe/pipeline definitions)
rm_outp = glob.glob("{}/*.dyne_output.hdf".format(path_ExpData))
rm_logs = glob.glob("{}/*.dyne_log.csv".format(path_ExpData))

for rm_type in [rm_outp, rm_logs]:
    for path in rm_type:
        os.remove(path)
```

```python
from multiprocessing import Pool
parallel_run = True

# Generate list of all pipe_defs and the pipeline architecture
proc_pdef = glob.glob("{}/*.dyne_pdef.json".format(path_ExpData))
proc_pline = "{}/ecog_coh_pipeline.json".format(path_ExpData)

# Setup helper function to map pipeline run
def _start_helper(path):
    print('\nProcessing: {}'.format(path))
    pline = dyne.pipeline.Pipeline(pipe_defs_json=path,
                                   pipeline_def_json=proc_pline)
    pline.start_pipeline()

if parallel_run:
    mp = Pool(15)
    mp.map(_start_helper, proc_pdef)
else:
    map(_start_helper, proc_pdef)
```

## Plot Adjacency Matrices

```python
%matplotlib inline
    
band = ['AlphaTheta', 'Beta', 'LowGamma', 'HighGamma']
sel_subj = np.random.permutation(subj_ids.keys())[0]
sel_name = 'HUP78-ictal-block-2' # np.random.permutation(subj_ids[sel_subj]['event_name'])[0]

for sel_band in band:
    # Get list of DyNe Output Files
    dyne_out = "{}/{}.{}.dyne_output.hdf".format(path_ExpData, sel_name, sel_band)
    dyne_log = "{}/{}.{}.dyne_log.csv".format(path_ExpData, sel_name,  sel_band)

    df_log = pd.read_csv(dyne_log, delimiter=',')
    pipe_hash = np.array(df_log[df_log.PIPE_NAME == 'MTCoh'].DOWNSTREAM_HASH)[0]

    df_outp = h5py.File(dyne_out, 'r')
    data = df_outp[pipe_hash]['data']

    # Set Parameters
    n_adj = data.shape[0]
    n_row = int(np.sqrt(n_adj))
    n_col = int(np.ceil(n_adj / n_row))

    # Figure
    plt.figure(figsize=(16, 12))
    for ii in xrange(n_adj):
        ax = plt.subplot(n_row, n_col, ii+1)
        ax.matshow(data[ii, :, :])
        ax.set_axis_off()
    plt.suptitle('{} - {}'.format(sel_name, sel_band))
```
