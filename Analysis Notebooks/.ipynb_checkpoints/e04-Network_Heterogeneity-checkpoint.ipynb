# Network Heterogeneity

## Initialize Environment

```python
try:
    %load_ext autoreload
    %autoreload 2
    
except:
    print 'NOT IPYTHON'

from __future__ import division
from IPython.display import display

import os
import sys
import glob
import h5py
import subprocess

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import rcParams

import fig_plotting
rcParams = fig_plotting.update_rcparams(rcParams)
#%matplotlib inline

os.chdir('../')
import Codebase
os.chdir('./Analysis Notebooks/')


path_CoreData = '/Users/akhambhati/Remotes/hoth_research/CoreData/IEEG_Neocortical'
path_PeriphData = '/Users/akhambhati/Remotes/hoth_research/PeriphData/ds-VCR_PushPull'
path_InpData = path_PeriphData + '/e01-Dyne_FuncNetw'
path_InpData_VR = path_PeriphData + '/e03-VirtualResection'
path_ExpData = path_PeriphData + '/e04-Heterogeneity'

for path in [path_CoreData, path_PeriphData, path_ExpData]:
    if not os.path.exists(path):
        print('Path: {}, does not exist'.format(path))
        os.makedirs(path)
        
display('Current directory: {}'.format(os.getcwd()))
```

## Generate List of Data

```python
# Get subject list
subj_event = [full_subj_path
              for full_subj_path in glob.iglob('{}/*.dyne_output.hdf'.format(path_InpData))]

subj_dict = {}
for path in subj_event:
    subj_id = path.split('/')[-1].split('.')[0].split('-')[0]
    epoch_id = path.split('/')[-1].split('.')[0].split('-')[1]
    event_id = subj_id + '-block-' + path.split('/')[-1].split('.')[0].split('-block-')[-1]
    epev_id = path.split('/')[-1].split('.')[0]
    band_id = path.split('/')[-1].split('.')[1]
    
    try:
        subj_dict[subj_id]
    except KeyError:
        subj_dict[subj_id] = {}

    try:
        subj_dict[subj_id][event_id]
    except KeyError:
        subj_dict[subj_id][event_id] = {}

    try:
        subj_dict[subj_id][event_id]['band_type']
    except KeyError:
        subj_dict[subj_id][event_id]['band_type'] = {}
        
    try:
        subj_dict[subj_id][event_id]['band_type'][band_id]
    except KeyError:
        subj_dict[subj_id][event_id]['band_type'][band_id] = {}
    
    subj_dict[subj_id][event_id]['band_type'][band_id][epoch_id] = \
               {'dyne_output': path,
                'dyne_log': '{}/{}.{}.dyne_log.csv'.format(path_InpData,
                                                           epev_id,
                                                           band_id),
                'centrality_path': '{}/{}.{}.centrality_distrib.npz'.format(path_InpData_VR, epev_id, band_id)
               }
```

## Generate List of Synchronizability, Node Strength Measure, Control Measure

```python
%matplotlib inline

df_dict = {'Subject_ID': [],
           'Event_ID': [],
           'Band_ID': [],
           'Epoch_ID': [],           
           'Time_ID': [],
           'Sync': [],
           'NStr_Mean': [],
           'NStr_Disp': [],
           'NStr_Skew': [],           
           'Contr_Mean': [],
           'Contr_Disp': [],
           'Contr_Skew': [],}

for subj_id in subj_dict.iterkeys():
    for event_id in subj_dict[subj_id].iterkeys():
        for band_id in subj_dict[subj_id][event_id]['band_type'].iterkeys():
            for epoch_id in subj_dict[subj_id][event_id]['band_type'][band_id].iterkeys():
            
                # Epoch
                f_dict = subj_dict[subj_id][event_id]['band_type'][band_id][epoch_id]

                # Get Synchronizability
                dyne_out = f_dict['dyne_output']
                dyne_log = f_dict['dyne_log']
                df_log = pd.read_csv(dyne_log, delimiter=',')
                pipe_hash_sync = np.array(df_log[df_log.PIPE_NAME == 'Synchronizability'].DOWNSTREAM_HASH)[0]
                df_outp = h5py.File(dyne_out, 'r')
                sync = df_outp[pipe_hash_sync]['data'][:, 0, 0]

                # Get Node Strength/Control Centrality
                cntr_out = np.load(f_dict['centrality_path'])
                ns = cntr_out['degree_centrality']
                cc = cntr_out['control_centrality']

                # Compute Measures
                ns /= ns.shape[1]
                ns_mean = ns.mean(axis=1)
                ns_disp = np.log10(np.std(ns, axis=1))
                ns_skew = stats.skew(ns, axis=1)
                
                cc_mean = cc.mean(axis=1)
                cc_disp = np.log10(np.std(cc, axis=1))
                #cc_skew = stats.skew(cc, axis=1)
                cc_skew = stats.skew(np.sign(cc), axis=1)                
                                
                
                for t_id, (ss, nsm, nsd, nsk, ccm, ccd, cck) in enumerate(
                    zip(sync, ns_mean, ns_disp, ns_skew, cc_mean, cc_disp, cc_skew)):
                    
                    df_dict['Subject_ID'].append(subj_id)
                    df_dict['Event_ID'].append(event_id)
                    df_dict['Band_ID'].append(band_id)
                    df_dict['Epoch_ID'].append(epoch_id)                
                    df_dict['Time_ID'].append(t_id)
                    df_dict['Sync'].append(ss)
                    df_dict['NStr_Mean'].append(nsm)
                    df_dict['NStr_Disp'].append(nsd)  
                    df_dict['NStr_Skew'].append(nsk)                      
                    df_dict['Contr_Mean'].append(ccm)
                    df_dict['Contr_Disp'].append(ccd)
                    df_dict['Contr_Skew'].append(cck)                    

df = pd.DataFrame(df_dict)                               
df.to_csv('{}/Population_NetworkHeterogeneity.csv'.format(path_ExpData))                
```

## Test Cases

### Synchronizability vs Node Strength Dispersion

```python
%matplotlib inline

df = pd.read_csv('{}/Population_NetworkHeterogeneity.csv'.format(path_ExpData))


# Plot
for band_sel in ['HighGamma']: #np.unique(df.Band_ID):
    df_sel = df[df.Band_ID == band_sel]
            
    plt.figure()
    ax = plt.subplot(111)
    print(len(df_sel.Sync))
    ax.scatter(df_sel.NStr_Disp, df_sel.Sync,
               color=[0.75, 0.75, 0.75], s=2.0)

    m, b, rho, pval, stderr = stats.linregress(df_sel.NStr_Disp, df_sel.Sync)
    ax.plot(df_sel.NStr_Disp, m*df_sel.NStr_Disp+b, color='k', alpha=0.75)
    print('{}: {}, {}'.format(band_sel, rho, pval))

    ax.set_title('{}'.format(band_sel))
    
    #ax.set_xlim(xmin=0)
    ticks = ax.get_xticks()
    ax.set_xticks(np.linspace(ticks[0], ticks[-1], 5))    
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xlabel('Node Strength Dispersion')    

    ax.set_ylim([0.0, 0.8])
    ticks = ax.get_yticks()
    ax.set_yticks(np.linspace(ticks[0], ticks[-1], 4))
    ax.yaxis.set_ticks_position('left')
    ax.set_ylabel('Synchronizability')
    
    plt.savefig('./e02-Figures/{}-NS_Disp-Sync.svg'.format(band_sel))
```

## Generate List of Control/Node Measures

```python
%matplotlib inline

df_dict = {'Subject_ID': [],
           'Event_ID': [],
           'Band_ID': [],
           'Epoch_ID': [],           
           'NStr': [],
           'Contr': []}

for subj_id in subj_dict.iterkeys():
    for event_id in subj_dict[subj_id].iterkeys():
        for band_id in subj_dict[subj_id][event_id]['band_type'].iterkeys():
            for epoch_id in subj_dict[subj_id][event_id]['band_type'][band_id].iterkeys():
            
                # Epoch
                f_dict = subj_dict[subj_id][event_id]['band_type'][band_id][epoch_id]

                # Get Synchronizability
                dyne_out = f_dict['dyne_output']
                dyne_log = f_dict['dyne_log']
                df_log = pd.read_csv(dyne_log, delimiter=',')
                pipe_hash_sync = np.array(df_log[df_log.PIPE_NAME == 'Synchronizability'].DOWNSTREAM_HASH)[0]
                df_outp = h5py.File(dyne_out, 'r')
                sync = df_outp[pipe_hash_sync]['data'][:, 0, 0]

                # Get Node Strength/Control Centrality
                cntr_out = np.load(f_dict['centrality_path'])
                ns = cntr_out['degree_centrality'] / cntr_out['degree_centrality'].shape[1]
                cc = cntr_out['control_centrality']
                
                ns = ns.reshape(-1)
                cc = cc.reshape(-1) 

                for i_id, (ns_single, cc_single) in enumerate(zip(ns, cc)):
                    df_dict['Subject_ID'].append(subj_id)
                    df_dict['Event_ID'].append(event_id)
                    df_dict['Band_ID'].append(band_id)
                    df_dict['Epoch_ID'].append(epoch_id)                
                    df_dict['NStr'].append(ns_single)
                    df_dict['Contr'].append(cc_single)                

df = pd.DataFrame(df_dict)                               
df.to_csv('{}/Population_ControlNovelty.csv'.format(path_ExpData))                
```
